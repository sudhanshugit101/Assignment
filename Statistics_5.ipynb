Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation statistics is a branch of statistics that involves estimating a population parameter based on a sample statistic. It is used when it is not feasible or practical to collect data on an entire population, and instead a smaller sample is taken from the population. 

Point estimate is a single value estimate of a population parameter based on a sample statistic.

Interval estimate, on the other hand, is a range of values that is used to estimate the unknown population parameter with a certain level of confidence. It is obtained by computing a confidence interval, which is a range of values that is likely to contain the true value of the population parameter with a specified level of confidence.

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    std_error = sample_std / math.sqrt(sample_size)
    margin_error = 1.96 * std_error
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    return (lower_bound, upper_bound)
```

Hypothesis testing is a statistical method used to make inferences about a population parameter based on sample data.

The importance of hypothesis testing lies in its ability to provide a systematic and objective approach to evaluating statistical evidence. It helps to eliminate biases and subjectivity in decision-making. Hypothesis testing also provides a framework for testing the validity of theories and models, and for making decisions based on data-driven evidence.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import numpy as np
from scipy.stats import ttest_ind

# generate two random samples
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=12, scale=2, size=50)

# calculate the mean and standard deviation of the samples
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# calculate the pooled standard deviation
pooled_std = np.sqrt(((len(sample1)-1)*(std1**2) + (len(sample2)-1)*(std2**2))/(len(sample1)+len(sample2)-2))

# conduct a two-sample t-test
t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=True)

# print the results
print("Sample 1 Mean: {:.2f}".format(mean1))
print("Sample 2 Mean: {:.2f}".format(mean2))
print("Pooled Standard Deviation: {:.2f}".format(pooled_std))
print("T-Statistic: {:.2f}".format(t_statistic))
print("P-Value: {:.4f}".format(p_value))
```

Q7: Write down the steps involved in hypothesis testing.

The general steps involved in hypothesis testing are as follows:

1. State the null hypothesis (H0) and alternative hypothesis (Ha)
2. Choose the appropriate test statistic and significance level
3. Collect data and calculate the test statistic
4. Calculate the p-value
5. Make a decision based on the p-value and the significance level:
6. If p-value < significance level, reject the null hypothesis
7. If p-value >= significance level, fail to reject the null hypothesis


Q8. Define p-value and explain its significance in hypothesis testing.

In hypothesis testing, the p-value is the probability of obtaining a test statistic at least as extreme as the observed results, assuming that the null hypothesis is true. It is used to determine the strength of evidence against the null hypothesis.

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 100)
plt.plot(x, t.pdf(x, df))
plt.title("Student's t-distribution with 10 Degrees of Freedom")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.show()
```

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy.stats import ttest_ind

# Generating two random samples of equal size
sample1 = np.random.normal(10, 2, 100)
sample2 = np.random.normal(12, 2, 100)

# Calculating the t-test for independent samples
t_stat, p_val = ttest_ind(sample1, sample2)

# Printing the results
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic is a measure of the difference between the sample mean and the population mean in terms of the standard error. It is used in hypothesis testing to determine whether the difference between two sample means is significant or due to chance.

The formula for the t-statistic is:

t = (x1 - x2) / (s(1/n1 + 1/n2))

Where x1 and x2 are the sample means, s is the pooled standard deviation, n1 and n2 are the sample sizes.

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

To test the hypothesis, we can perform a one-sample t-test with the following null and alternative hypotheses:

Null hypothesis: The true mean decrease in blood pressure is not significantly different from 10 mmHg.
Alternative hypothesis: The true mean decrease in blood pressure is significantly different from 10 mmHg.

We can use the following formula to calculate the t-statistic:

t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))

Substituting the values given in the question, we get:

t = (8 - 10) / (3 / sqrt(100)) = -4.47

Using a significance level of 0.05 and 99 degrees of freedom (sample size minus one), we can find the critical t-value using a t-table or a t-distribution calculator. The critical t-value for a two-tailed test with a significance level of 0.05 and 99 degrees of freedom is approximately 1.984.

Since the absolute value of the calculated t-statistic (4.47) is greater than the critical t-value (1.984), we can reject the null hypothesis and conclude that the true mean decrease in blood pressure is significantly different from 10 mmHg at a significance level of 0.05.

t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))
t = (4.8 - 5) / (0.5 / sqrt(25))
t = -2

Using a t-table with 24 degrees of freedom (n-1), we can find the critical t-value for a one-tailed test at a 0.01 significance level. The critical t-value is -2.492.

Since our calculated t-value (-2) is greater than the critical t-value (-2.492), we reject the null hypothesis. Therefore, we can conclude that the true mean weight of the products is less than 5 pounds at a significance level of 0.01.
